```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/2023_aicoursework_llm/datas/kdrama.csv")
filter_df = df.drop(["Aired Date","Aired On","Duration","Content Rating","Production companies", "Rank"], axis=1)
filter_df
```

```python
filter_df.info()
```

```python
import chromadb
client = chromadb.PersistentClient()

collection = client.get_or_create_collection(
    name="k-drama",
    metadata={"hnsw:space": "cosine"}# l2 is the default
)
```

```python

# 데이터 인덱스
ids = []
# 메타데이터
doc_meta = []
# 벡터로 변환 저장할 텍스트 데이터로 ChromaDB에 Embedding 데이터가 없으면 자동으로 벡터로 변환해서 저장
documents = []

for idx in range(len(filter_df)):
    item = filter_df.iloc[idx]
    id = item['Name'].lower().replace(' ','-')

    document = f"{item['Name']} : {item['Synopsis']} : {str(item['Cast']).strip().lower()} : {str(item['Genre']).strip().lower()}"
    meta = {
        "rating" : item['Rating']
    }

    ids.append(id)
    doc_meta.append(meta)
    documents.append(document)

# DB 저장
collection.add(
    documents=documents,
    metadatas=doc_meta,
    ids=ids
)
# DB 쿼리
collection.query(
    query_texts=["romantic comedy drama"],
    n_results=5,
)
```

```python
import json
import openai
openai.api_key = ""
```

```python
def recommend_drama(query, n=10):
  sim_list = collection.query(
      query_texts=[query],
      n_results=n,
  )

  documents = sim_list["documents"][0]

  drama_list = []

  for doc in documents:
    doc_data = doc.split(" : ")
    drama_list.append(
      {
        "name": doc_data[0],
        "synopsis": doc_data[0],
        "cast": doc_data[0],
        "genre": doc_data[0],
      }
    )

  return json.dumps(drama_list)

dramas = recommend_drama("korean drama", n=2)
dramas
```

```python
def recommend_drama_openai(query, n=10):
  messages = [
      {
          "role": "user",
          "content": f"{query} 추천 개수는 {n}개로 해줘."
          }
      ]
  functions = [
      {
            "name": "recommend_drama",
            "description": "드라마를 추천해줍니다.",
            "parameters": {
                "type": "object",
                "properties": {
                    "query": {
                        "type": "string",
                        "description": "추천 할 드라마 종류",
                    },
                    "n": {
                        "type": "integer",
                        "description": "추천 개수",
                    },
                },
                "required": ["query", "n"],
            },
        }
  ]

  response = openai.ChatCompletion.create(
      model="gpt-3.5-turbo",
      messages=messages,
      functions=functions,
      function_call="auto"
      )

  response_message = response["choices"][0]["message"]
  if response_message.get("function_call"):
    # Note: the JSON response may not always be valid; be sure to handle errors
    available_functions = {
        "recommend_drama": recommend_drama,
    }
    function_name = response_message["function_call"]["name"]
    fuction_to_call = available_functions[function_name]
    function_args = json.loads(response_message["function_call"]["arguments"])
    function_response = fuction_to_call(
        query=function_args.get("query"),
        n=function_args.get("n"),
    )

    messages.append(response_message)
    messages.append(
        {
            "role": "function",
            "name": "recommend_drama",
            "content": function_response,
        }
    )
    messages.append(
        {
            "role": "user",
            "content": "데이터를 활용하여 드라마를 추천해줘. 결과에는 name, cast, genre 가 포함되어야 해"
        }
    )

    second_response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo-0613",
        messages=messages,
    )


    json_data = json.dumps(second_response, ensure_ascii=False)

    return second_response.choices[0].message.content
```
