```python
def chat_completion(prompt, temperature=0, max_tokens=512):
  completion = openai.ChatCompletion.create(
      model="gpt-3.5-turbo",
      messages=[
        {
            "role": "user",
            "content": prompt
            }
    ],
      temperature=temperature,
      max_tokens=max_tokens,
      )

  result = completion.choices[0].message.content

  return result

def get_prompt(target, skin_type):
  prompt = f"""
  ₩₩₩
  'skin type': {skin_type},
  'target': {target}
  ₩₩₩

  위에서 '₩₩₩' 으로 구분된 입력값에 대해서 아래 내용을 수행해줘.
  'skin type' 을 가진 'target' 이 피부 관리를 할 때 주의해야 할 점에 대한 블로그 글을 500단어 이상 600단어 이하로 작성해줘.
  글은 서론, 본론, 결론으로 구분해줘.
  본론은 피해야 할 것과 사용해아 할 것으로 구분하고, 하위 넘버링을 해줘.

  출력은 아래 [출력 예시] 와 같은 포맷을 가진 마크다운으로 만들어줘. 레이블은 제목으로, 내용은 본문으로 만들어줘. 하위 넘버링은 하위 레이블링으로 표시해줘.

  [출력 예시]
  # 제목
  ## 서론
  ## 본론
  ### 1. 피해야 할 것
  #### 1.1
  #### ...
  ### 2. 사용해야 할 것
  #### 2.1.
  #### ...
  ## 결론
  """

  return prompt



```

```python


target_list = ['20대 여성']#, '20대 남성', '30대 여성','30대 남성']
skin_type_list = ['지성 피부']#,'건성 피부']
temperature = 0
max_tokens = 3000

for target in target_list:
  for skin_type in skin_type_list:
    prompt = get_prompt(target, skin_type)
    print(f"target: {target}")
    print(f"skin type: {skin_type}")
    print("prompt")
    print(prompt)
    print()

    result = chat_completion(prompt, temperature=temperature, max_tokens=max_tokens)

    print("================== chat gpt result ==================")
    print("length: ", len(result))
    print("word num: ", len(result.split(" ")), "\n")
    print(result)
    print()
    break
```
